Show the simple example of how the mean correction can improve phase-averaging accuracy.


```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
from functions.timestepping import *
```

```python
def original_ode_sol(t, c, eps, omega, u0):
    return (1j*c*eps/omega)*(np.exp(-1j*omega*t/eps)-1) + np.exp(-1j*omega*t/eps)*u0

def modvar_ode_sol(t,c,eps,omega,u0):
    return u0 + eps*c/(1j*omega)*(np.exp(1j*omega*t/eps) - 1)

#def standard_phase_averaged(t,c,eps,omega,u0):
```

```python
t = np.linspace(0,10,1000)

u0 = 1 + 1j
eps = 0.1
omega = 1

# Use a constant nonlinearity, or a slowly varying one
c = 1
```

```python
analyt_sol = original_ode_sol(t, c, eps, omega, u0)
v_sol = modvar_ode_sol(t, c, eps, omega, u0)
```

```python
plt.figure()
plt.plot(t, np.real(analyt_sol),label='real')
plt.plot(t, np.imag(analyt_sol),label='imag')
plt.plot(t, np.abs(analyt_sol),label='abs)
#plt.plot(t,c)
```

```python
plt.figure()
plt.plot(t,np.real(v_sol),label='real')
plt.plot(t, np.imag(v_sol), label='imag')
plt.plot(t, np.abs(v_sol), label='abs')
plt.legend()
```

```python
plt.figure()
plt.plot(t,analyt_sol, label='Original space solution')
plt.plot(t, v_sol, label='Modulation variable solution')
plt.legend(loc = 'lower center')
plt.ylabel('u/v')
plt.xlabel('Time')
```
